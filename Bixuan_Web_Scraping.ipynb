```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
#load a page
response = requests.get("https://www.yellowpages.com/search?search_terms=coffee&geo_location_terms=Washington%2C+DC")
```

```python
response
```

```python
#display contents of response
response.content
```

```python
#declare a soup variable and read in the content
soup = BeautifulSoup(response.content, 'lxml')
```

```python
print(soup)
```

```python
#prettify the results
print(soup.prettify())
```

```python
#list all the anchor tags (<a>)
soup.find_all("a")
```

```python
#find all the <h1> tags
soup.find_all("h1")
```

```python
#show all the links
for link in soup.find_all("a"):
    print(link)
    print('****************************************************')
```

```python
#show all the hrefs
for link in soup.find_all("a"):
    print(link.get("href"))
```

```python
#print all the text
for link in soup.find_all("a"):
    print(link.text)
```

```python
#print all the text and all the href
for link in soup.find_all("a"):
    print(link.text, link.get("href"))
```

```python
for link in soup.find_all("a"):
    print("<a href='%s'>%s</a>" %(link.get('href'), link.text))
```

```python
g_data = soup.find_all("div", attrs={"class":"info"})
```

```python
g_data
```

```python
#clean up the results a bit
for item in g_data:
    print(item.text)
```

```python
#get the items as a list
for item in g_data:
    print(item.contents)
```

```python
#get the first two items of a list
items_list = []
for item in g_data:
    items_list.append(item)
items_list[0:2]
```

```python
#print the 0th element content and text
for item in g_data:
    print(item.contents[0])

for item in g_data:
    print(item.contents[0].text)
```

```python
#get the name and addresses of the places
for item in g_data:
    print(item.contents[0].text)
    print(item.contents[1].text)
```

```python
#get more specific using the class abd build a list
for item in g_data:
    print(item.contents[0].find_all("a", attrs={"class":"business-name"})[0])
```

```python
# make a pretty list of locations, names, and addresses
for item in g_data:
    if item.contents[1].find_all("span")[0].attrs == {'class': ['count']}:
        print(item.contents[0].find_all("a",attrs={"class":"business-name"})[0].text)
        print(item.contents[1].find_all('span',attrs={"class":"street-address"})[0].text,
              item.contents[1].find_all("span",attrs={"class":"locality"})[0].text,
              item.contents[1].find_all("span")[3].text,
              item.contents[1].find_all("span")[4].text)
        print(item.contents[1].find_all('div',attrs={"class":"phones phone primary"})[0].text)
    else:
        print(item.contents[0].find_all("a",attrs={"class":"business-name"})[0].text)
        print(item.contents[1].find_all('span',attrs={"class":"street-address"})[0].text,
              item.contents[1].find_all("span",attrs={"class":"locality"})[0].text,
              item.contents[1].find_all("span")[2].text,
              item.contents[1].find_all("span")[3].text)
        print(item.contents[1].find_all('div',attrs={"class":"phones phone primary"})[0].text)
    print()
```

```python
#make a list for business name
business_name=[]
for item in g_data:
    business_name.append(item.contents[0].find_all("a", attrs={"class":"business-name"})[0].text)
```

```python
#make a list for street address
street_address=[]
for item in g_data:
    street_address.append(item.contents[1].find_all('span',attrs={"class":"street-address"})[0].text)                          
```

```python
#make a list for region 
region=[]
for item in g_data:
    if item.contents[1].find_all("span")[0].attrs == {'class': ['count']}:
        region.append(str(item.contents[1].find_all("span",attrs={"class":"locality"})[0].text.strip()+" "+item.contents[1].find_all("span")[3].text))
    else:
        region.append(item.contents[1].find_all("span",attrs={"class":"locality"})[0].text.strip()+" "+item.contents[1].find_all("span")[2].text)
```

```python
#make a list for postcode
postcode=[]
for item in g_data:
    if item.contents[1].find_all("span")[0].attrs == {'class': ['count']}:
        postcode.append(item.contents[1].find_all("span")[3].text+" "+item.contents[1].find_all("span")[4].text)
    else:
        postcode.append(item.contents[1].find_all("span")[2].text+" "+item.contents[1].find_all("span")[3].text)
```

```python
#make a list for phone
phone=[]
for item in g_data:
    phone.append(item.contents[1].find_all('div',attrs={"class":"phones phone primary"})[0].text)
```

```python
#make a dataframe of locations, names, and addresses
info_df = pd.DataFrame({"business_name":business_name, 
                       "street_address":street_address,
                       "region":region,
                       "postcode":postcode,
                       "phone":phone})
                      
info_df
```
